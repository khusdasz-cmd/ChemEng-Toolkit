# FT-IR Spectroscopy Analysis Example
## Fractionated Lignin FT-IR Spectra

Load and visualize FT-IR spectra with peak assignment annotations.

```python
import sys
sys.path.insert(0, "..")
from chemeng_toolkit.thermal_analysis import FTIRProcessor
```

## 1. Load FT-IR Data

```python
processor = FTIRProcessor(data_dir="../data/ftir_sample")
files = [
    ("MBL-5-10K.CSV", "L5", "#1f77b4"),
    ("MBL-10-20K.CSV", "L10", "#ff7f0e"),
    ("MBL-20-100K.CSV", "L20", "#2ca02c"),
    ("gt_100k.CSV", "L100", "#d62728"),
]
processor.load_batch(files)
```

## 2. Stacked FT-IR Spectra

```python
processor.plot_stacked(title="FT-IR Spectra of Fractionated Lignin", show=True)
```

## 3. Normalized Spectra

```python
processor.plot_stacked(normalize=True, title="FT-IR Spectra (Normalized)", show=True)
```

## 4. Individual Subplots

```python
processor.plot_individual(show=True)
```
